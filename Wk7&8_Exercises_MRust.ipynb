```python
# Import necessary libraries
import numpy as np
import pandas as pd
```

```python
# I'm going to use my candy df from chapter 7 that has assigned values to Joy, Meh, and Despair
candy.head()
```

```python
# Clean up column names a little bit
cleaner_candy = candy.rename(columns=lambda x: x.strip().lower().replace('[', '').replace(']', ''))
cleaner_candy.head()
```

```python
# Subset a small piece of the data to work with
piece_of_candy = cleaner_candy[['butterfinger', '100 grand bar', 
                               'anonymous brown globs that come in black and orange wrappers', 
                               'any full-sized candy bar', 'black jacks', 'bonkers', 'bottle caps']]
piece_of_candy.head()
```

```python
# Create dictionary mapping
mapping = {'butterfinger': 'chocolate', '100 grand bar': 'chocolate', 
           'anonymous brown globs that come in black and orange wrappers': 'chews', 
           'any full-sized candy bar': 'chocolate', 'black jacks': 'hard candy', 'bonkers': 'hard candy',
           'bottle caps': 'hard candy'}
```

```python
# Group by the dictionary mapping 
by_types = piece_of_candy.groupby(mapping, axis=1)
```

```python
# Sum up the total of scores by type for each observation
by_types.sum()
```

```python
piece_of_candy.groupby(mapping, axis=1).describe()
```

```python
# As we can see, there are some Nan values in our dataset. I'm using the means listed in the describe stats above to 
# estimate what the mode for that group type would be.
piece_of_candy
```

```python
# I am going to use split/apply/combine to fill in the NaNs of each candy with the estimated mode.
fill_values = {'chews': 1, 'chocolate': 5, 'hard candy': 3}
fill_func = lambda g: g.fillna(fill_values[g.name])
piece_of_candy.groupby(mapping, axis=1).apply(fill_func)
```

```python
# Choosing to use the candy dataset for this since there is a timestamp
candy = pd.read_excel("CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx")
candy.dropna(axis=1, how='all', inplace=True)
candy.replace({'JOY': 5, 'MEH': 3, 'DESPAIR': 1}, inplace=True)
candy.head()
```

```python
# Change the column into a datetime
from datetime import datetime
pd.to_datetime(candy['Timestamp'])
```

```python
# I'm going to groupby and offset the data to the month end to calculate the means. To do this I have to import 
# the MonthEnd from the pandas tseries library
from pandas.tseries.offsets import MonthEnd
```

```python
offset = MonthEnd()
candy.groupby(offset.rollforward).mean()
```

```python
# After messing with that forever, I'm going to take a different route. I'm going to generate a date range and add it
# to the metobs dataframe. 
# Read in file using pandas
metobs = pd.read_excel("metobs.xlsx")
metobs
```

```python
# We see that metobs is 10832 rows. I will create a date index to merge with the dataframe.
metobs['DateTime'] = pd.date_range(start='2012-04-01', periods=10832, freq='H')
metobs.head()
```

```python
# Let's make the DateTime column the index. 
metobs.set_index('DateTime', drop=False, inplace=True)
metobs.head()
```

```python
# Now let's try the whole offset thing again, but since we have no good data to find the mean of, how about we count
# the number of observations in the groups. 
offset = MonthEnd()
metobs.groupby(offset.rollforward).count()
```

```python
# Oh, I get it. The offset worked, but I had added the hourly in there and it only rolled forward the day to the end 
# month, keeping the hourly inputs. But it did count the observations in each column for that month. Cool!
```
